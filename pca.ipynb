# PCA - Principal Component Analysis
Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables into a set of values of linearly uncorrelated variables called principal components.

## Creating synthetic data

Create 3D data with two dimensions sampled from a Normal distribution and one dimension as the sum of the first two.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

x1 = np.random.normal(size=100)
x2 = np.random.normal(size=100)
x3 = x1 + x2
X = np.c_[x1, x2, x3]

# show the first 5 data points
print X[1:5,:]
```

## Show the data

```python
fig = plt.figure()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

ax.scatter(x1, x2, x3)
```

## Apply PCA
Apply the Principal Component Analysis to the data and show the explained variance for each of dimensions.

```python
pca = PCA(svd_solver='full')
pca.fit(X)

print(pca.explained_variance_)
```

## Show results
Apply the dimensionality reduction on X keeping only the significant components.

```python
pca.n_components = 2
X_reduced = pca.fit_transform(X)

plt.scatter(X_reduced[:,0], X_reduced[:,1])
```
